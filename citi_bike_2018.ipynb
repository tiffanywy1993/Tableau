```python
import pandas as pd
```

```python
filenames=['Resources/201801-citibike-tripdata.csv','Resources/201802-citibike-tripdata.csv','Resources/201803-citibike-tripdata.csv','Resources/201804-citibike-tripdata.csv','Resources/201805-citibike-tripdata.csv','Resources/201806-citibike-tripdata.csv','Resources/201807-citibike-tripdata.csv','Resources/201808-citibike-tripdata.csv','Resources/201809-citibike-tripdata.csv','Resources/201810-citibike-tripdata.csv','Resources/201811-citibike-tripdata.csv','Resources/201812-citibike-tripdata.csv']
```

```python
dataframes=[]
for file in filenames:
    dataframes.append(pd.read_csv(file))
```

```python
citibike_2018=pd.concat(dataframes, axis=0, ignore_index=True)
citibike_2018.head()
```

```python
citibike_2018=citibike_2018.sample(n=500000)
```
